```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a 2x2 grid of subplots with shared x and y axes
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(12, 8))

# Adjust the layout automatically
plt.tight_layout()

# Plot different data on each subplot
axes[0, 0].plot(x, y1, 'r-*')
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, y2, 'b-o')
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].plot(x, y1 + y2, 'g-')
axes[1, 0].set_title('Sine + Cosine')

axes[1, 1].scatter(x, y1 * y2)
axes[1, 1].set_title('Sine * Cosine')

# Set labels
for ax in axes.flat:
    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a new figure and add axes
fig = plt.figure(figsize=(8, 6))
main_axes = fig.add_axes([0, 0, 1, 1])

# Plot the data with labels
main_axes.plot(x, y1, 'r-*', label='Sine Wave')
main_axes.plot(x, y2, 'b-o', label='Cosine Wave')

# Set labels and title
main_axes.set_xlabel('X Axis')
main_axes.set_ylabel('Y Axis')
main_axes.set_title('Main Plot')

# Add the legend
main_axes.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
x = np.linspace(0, 10, 100)
z = np.exp(x)

# Create a new figure
fig = plt.figure(figsize=(8, 6))

# Add the main set of axes occupying the entire figure
axes_main = fig.add_axes([0, 0, 1, 1])
axes_main.plot(x, z, 'r-*', label='Exponential Growth')
axes_main.set_xlabel('X Axis')
axes_main.set_ylabel('Y Axis')
axes_main.set_title('Main Plot')
axes_main.legend()

# Add a smaller set of axes at a specific position
axes_inset = fig.add_axes([0.1, 0.6, 0.2, 0.2])
axes_inset.plot(x, z, 'b-o', label='Exponential Growth')
axes_inset.set_xlabel('X Axis')
axes_inset.set_ylabel('Y Axis')
axes_inset.set_title('Inset Plot')
axes_inset.set_xlim(1, 2)
axes_inset.set_ylim(0, 5)
axes_inset.legend()

# Show the plot
plt.show()
```
