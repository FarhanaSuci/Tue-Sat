```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame (same as before)
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68],
    'tip': [1.01, 1.66, 3.50, 3.31],
    'smoker': ['No', 'No', 'No', 'Yes'],
    'day': ['Sun', 'Sun', 'Sun', 'Sun'],
    'time': ['Dinner', 'Dinner', 'Dinner', 'Dinner'],
    'size': [2, 3, 3, 2]
}

tips = pd.DataFrame(data)

# Select only numeric columns for correlation calculation
numeric_cols = tips.select_dtypes(include='number').columns
tc = tips[numeric_cols].corr()

# Create a heatmap with adjusted figure size and font size
plt.figure(figsize=(10, 8))  # Adjust the figure size
sns.heatmap(tc, annot=True, annot_kws={"size": 10})  # Adjust the font size for annotations

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for correlation matrix
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68],
    'tip': [1.01, 1.66, 3.50, 3.31],
    'size': [2, 3, 3, 2]
}

tips = pd.DataFrame(data)

# Compute the correlation matrix
tc = tips.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))  # Increase figure size if needed

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(tc, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)

# Show the plot
plt.show()
```
