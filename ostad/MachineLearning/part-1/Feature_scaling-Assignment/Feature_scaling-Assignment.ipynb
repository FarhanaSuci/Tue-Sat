### Read data from income.csv and scale the relevant features using standardization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('income.csv')
```

```python
df
```

```python
X=df[['Age','Income']]
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_standarization=sc.fit_transform(X)
```

```python
X_standarization
```

```python
X_stand=X.copy()
```

```python
X_stand['Age']=(X_stand['Age']-X_stand['Age'].mean())/X_stand['Age'].std()
```
